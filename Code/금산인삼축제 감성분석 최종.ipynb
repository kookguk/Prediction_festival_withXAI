```python
import json
import pandas as pd

df = pd.read_excel('금산인삼축제 워드클라우드, 감성분석 병합.xlsx')
df
```

```python
# 문자열을 딕셔너리로 변환하는 함수
def str_to_dict(conf_str):
    # 작은 따옴표를 큰 따옴표로 변경 (json.loads가 제대로 작동하게 하기 위해)
    conf_str = conf_str.replace("'", '"')
    return json.loads(conf_str)

# confidence 칼럼의 문자열을 딕셔너리로 변환
df['confidence'] = df['confidence'].apply(str_to_dict)

print(df)
```

```python
# 딕셔너리를 별도의 칼럼으로 분리
confidence_df = df['confidence'].apply(pd.Series)

# 반올림하여 소수점 첫번째 자리까지 표시
confidence_df = confidence_df.round(2)

# 원본 데이터프레임에 새로운 칼럼 추가
df = pd.concat([df.drop(columns=['confidence']), confidence_df], axis=1)

print(df)
```

```python
# 결과를 엑셀 파일로 저장
df.to_excel('금산인삼축제 감성분석 최종.xlsx', index=False)
```
