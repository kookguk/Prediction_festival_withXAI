## 1. 제목 칼럼 전처리

```python
df = data[['title']]
```

```python
for i in df.title:
    print(i)
```

```python
import numpy as np
df['title'] = df['title'].astype(str)
#문자열 함수를 적용하기 때문에 객체값을 모두 str로 변환
```

```python
import string

#특수문자 전처리 함수 
def str_punc(x):
    pre_str=[]
    for i in x:
        if i not in string.punctuation:
            pre_str.append(i)
    pre_str = ''.join(pre_str)
    return pre_str
```

```python
#apply 함수를 전체에 적용한다.
df['title'] = df['title'].apply(str_punc) 
```

```python
# 반복문과 정규식을 통해 전처리
import re
list_reg = []
for i in df.title:
    if isinstance(i, str):  #데이터가 문자열인지 한 번 더 확인하여 예외처리
        title = re.sub('[^가-힣]',' ',i).strip()
        if (title !=''):
            list_reg.append(title)
```

```python
list_reg
```

## 2. 제목 칼럼 형태소 분석

```python
from konlpy.tag import Okt

okt = Okt()

# 사용자 정의 불용어 사전
custom_stop = set(['는', '은', '이', '가', '을', '도', '를', '등'])

# 텍스트 토큰화 (명사, 동사, 형용사)
tokenized_text = []
for text in list_reg:
    pos_tokens = okt.pos(text)
    filtered_tokens = [word for word, pos in pos_tokens if pos in ['Noun', 'Verb', 'Adjective']]
    tokenized_text.append(filtered_tokens)

# 불용어 제거
filtered_text = []
for tokens in tokenized_text:
    filtered_tokens = [token for token in tokens if token not in custom_stop]
    filtered_text.append(" ".join(filtered_tokens))

# 결과 확인
for text in filtered_text:
    print(text)
```

```python
#2차원 리스트를 1차원 리스트로 변환
tk_sp = [i for j in tokenized_text for i in j]
```

```python
tk_sp
```

## 3. 텍스트 카운팅

```python
# 위의 추출한 형태소들을 실제 카운팅
sp_cnt = dict()
for word in tk_sp:
    sp_cnt[word] = sp_cnt.get(word, 0)+1

sp_cnt
```

```python
df_last = pd.DataFrame(list(sp_cnt.items()), columns=['text', 'count'])

print(df_last)
```
